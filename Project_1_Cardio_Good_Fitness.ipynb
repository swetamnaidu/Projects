1.3.6 Checking for missing values

```python
def missing_check(df):
    total = df.isnull().sum().sort_values(ascending=False)   # total number of null values
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)  # percentage of values that are null
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])  # putting the above two together
    return missing_data # return the dataframe
missing_check(data)
```

1.6.3 Correlation between Marital Status, Gender and  Usage

```python
product_usage = data['Usage'].groupby([data['MaritalStatus'],data['Gender']]).sum()  # number of suicides by country and sex
product_usage = product_usage.reset_index().sort_values(by='Usage',ascending=False) # sort in descending order


fig = plt.figure(figsize=(15,5))
plt.title('Product Usage')

sns.barplot(y='MaritalStatus',x='Usage',hue='Gender',data = product_usage, palette='Set2');

plt.ylabel('Customer Atrribute')
plt.tight_layout()
```

<font color=darkblue>
<font size =3>
# Observations 

<font color = 'darkblue'>

Product Usage of partnered customers is more than single customers. And more male customers use the treadmill than female customers
    

<font color=darkorange>

1.6.4 Correlation between Product, Age and  Usage

```python
product_usage = data['Usage'].groupby([data['Age'],data['Product']]).sum()  # number of suicides by country and sex
product_usage = product_usage.reset_index().sort_values(by='Age',ascending=False) # sort in descending order


fig = plt.figure(figsize=(15,5))
plt.title('Product Usage as per age')

sns.lineplot(y='Age',x='Usage',hue='Product',data = product_usage, palette='Set2');

plt.ylabel('Age')
plt.tight_layout()
```
